```python
from surprise import SVD
from surprise import Dataset, dataset
from surprise import evaluate, print_perf
```

```python
# Loading the dataset and divide it into 3 for cross-validation
filePath = '/Users/JeongSooMin/Desktop/Gpaper/dataset/ml-latest-small/ratings.csv'
reader = dataset.Reader(line_format='user item rating timestamp', sep=',')
```

```python
data = Dataset.load_from_file(filePath, reader=reader)
```

```python
data.split(n_folds=20)

# We'll use the famous SVD algorithm.
algo = SVD()

# Evaluate performances of our algorithm on the dataset.
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])

print_perf(perf)
```
